# Collect Additional Data for Cleaned Dataset

```python
# Import Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import os
from config import gkey
```

### Pull in the correct file

```python
ticker_excel = pd.read_excel('Tickers_to_Include_in_Database.xlsx')
```

```python
ticker_df = pd.DataFrame(ticker_excel)
ticker_df
```

### Set up empty arrays for data that is being collected

```python
performance_2019 = []
market_cap_as_of_Jan_30 = []
exchange = []
ceo = []
logo = []
website = []
industry = []
sector = []
description = []
profile_ticker = []
i = 1
```

### Run "For Loop" to pull data from API for each of the tickers

```python
for ticker in ticker_df['Ticker']:
    query_url = f"https://financialmodelingprep.com/api/v3/company/profile/{ticker}"
    ticker_response = requests.get(query_url)
    response_json = ticker_response.json()
    profile_json = response_json['profile']
    market_cap_as_of_Jan_30.append(profile_json['mktCap'])
    exchange.append(profile_json['exchange'])
    ceo.append(profile_json['ceo'])
    logo.append(profile_json['image'])
    website.append(profile_json['website'])
    industry.append(profile_json['industry'])
    sector.append(profile_json['sector'])
    description.append(profile_json['description'])
    performance_url = f"https://financialmodelingprep.com/api/v3/historical-price-full/{ticker}?from=2018-12-31&to=2019-12-31"
    performance_response = requests.get(performance_url)
    performance_json = performance_response.json()
    performance_2019.append(((performance_json['historical'][-1]['close']/performance_json['historical'][0]['close'])-1)*100)   
    profile_ticker.append(ticker)
    print(f"Finished Company #: {i}")
    i =i + 1
```

### Pull the columns you want from the original data frame and begin to add columns you pulled

```python
ESG_df = pd.DataFrame(ticker_df[['Ticker','Company Name','TR_HQ_City','TR_HQ_State/Province','TR_HQ_Country','ESG Risk Score','ESG Risk Score Level','ESG Risk Score Percentile','Environment Risk Score','Social Risk Score','Governance Risk Score','Controversy Level Risk Text']])
ESG_df.columns = ['Ticker','Company Name','HQ City','HQ State/Province','HQ Country','ESG Risk Score','ESG Risk Score Level','ESG Risk Score Percentile','Environment Risk Score','Social Risk Score','Governance Risk Score','Controversy Level Risk Text']
ESG_df['2019 Performance'] = performance_2019
ESG_df['Market Cap'] = market_cap_as_of_Jan_30
ESG_df['Exchange'] = exchange
ESG_df['CEO'] = ceo
ESG_df['Logo Image URL'] = logo
ESG_df['Website URL'] = website
ESG_df['Industry'] = industry
ESG_df['Sector'] = sector
ESG_df['Company Description'] = description
```

### If you need to reformat any columns, do it here

```python
ESG_df['HQ City'] = ESG_df['HQ City'].str.capitalize()
ESG_df['HQ State/Province'] = ESG_df['HQ State/Province'].str.capitalize()
ESG_df['HQ Country'] = ESG_df['HQ Country'].str.capitalize()
```

```python
ESG_df
```

### Collect Latitude & Longitude

```python
ESG_df["Latitude"] = ""
ESG_df["Longitude"] = ""
ESG_df["address"] = ESG_df['HQ City'] + "," + ESG_df['HQ State/Province'] + "," + ESG_df['HQ Country']
ESG_df.address
```

```python
# create a params dict that will be updated with new city each iteration
params = {"key": gkey}

# Loop through the cities_pd and run a lat/long search for each city
for i, row in esg_df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    address = row['address']
    

    # update address key value
    params['address'] = f"{address}"
    
    try:
    # make request
        cities_lat_lng = requests.get(base_url, params=params)
    
    #print the cities_lat_lng url, avoid doing for public github repos in order to avoid exposing key
        print(cities_lat_lng.url)
    
    # convert to json
        cities_lat_lng = cities_lat_lng.json()

        esg_df.loc[i, "Latitude"] = cities_lat_lng["results"][0]["geometry"]["location"]["lat"]
        esg_df.loc[i, "Longitude"] = cities_lat_lng["results"][0]["geometry"]["location"]["lng"]
        
    except:
        print("Adress not found...skipping")
        pass

# Indicate that Data Loading is complete 
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```
